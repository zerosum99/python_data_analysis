## 데이터 읽기

- 깃허브는 하나의 파일을 100메가 이상 저장하지 못해서 두 개의 파일로 분리


```python
import pandas as pd
```

```python
df_1 = pd.read_csv('./data/creditcard_1.csv')
```

```python
df_1.head()
```

## 3-3 데이터 전처리

### 의미없는 칼럼 삭제

```python
df_1 = df_1.drop('Unnamed: 0', axis=1)
```

```python
df_2 = pd.read_csv('./data/creditcard_2.csv')
```

```python
df_2.head()
```

```python
df_2 = df_2.drop('Unnamed: 0', axis=1)
```

## 두 개의 파일을 하나로 합친다.

- 축이 방향은 수직으로 처리하므로 axis=0을 지정

```python
df = pd.concat([df_1, df_2], axis=0)
```

```python
df.shape
```

## 3-4 데이터 분리

### 모델에 대한 평가 함수 

```python
from sklearn.metrics import classification_report, accuracy_score
```

### 데이터를 훈련과 테스트 세트로 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['Time','Class'], axis=1), 
                                                    df['Class'], test_size=0.3, random_state=0)
```

## 3-5 부스팅 알고리즘으로 확인

```python
from xgboost import XGBClassifier
```

## 모델 생성

```python
xgb = XGBClassifier()
```

## 훈련하기

```python
xgb.fit(X_train, Y_train)
```

## 예측하기

```python
y_pred = xgb.predict(X_test)
```

## 평가하기

```python
n_errors = (y_pred != Y_test).sum()
# Run Classification Metrics
print("{}: {}".format('xgboost',n_errors))
print("Accuracy Score :")
print(accuracy_score(Y_test,y_pred))
print("Classification Report :")
print(classification_report(Y_test,y_pred))
```

## 텐서플로우 모듈 사용

```python
import tensorflow as tf
```

## 딥러닝으로 처리할 때는 모든 칼럼을 정규화나 표준화 필요


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
```

### 테스트 데이터도 정규화는 훈련데이터를 기준으로 처리한다

```python
X_test = scaler.transform(X_test)
```

## 하나의 히든과 출력 레이어 지정

- 과적합을 방지하기 위해 드롭아웃 레이어 추가
- 히든 레이어의 결과값 중에 30%는 출력레이어로 정보가 전달되지 않는다


```python
model = tf. keras.Sequential([
                               tf.keras.layers.Dense(32, activation='relu',input_shape=(X_train.shape[-1],)),
                               tf.keras.layers.Dropout(0.3),
                               tf.keras.layers.Dense(1, activation='sigmoid'),
                      ])
```

## 모델에 대한 손실함수, 정규화 추가

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=1e-3), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

## 모델 훈련 

```python
model.fit(X_train, Y_train, batch_size=32, epochs=20)
```

## 모델 테스트 및 평가

```python
model.evaluate(X_test, Y_test)
```

### 테스트 결과를 반환 

```python
y_pred_tf = model.predict(X_test)
```

```python
y_pred_tf.shape
```

## 분류 평가를 위한 혼동메트릭스 생성 

- 사기로 예측은 115 + 12 = 127건


```python
confusion_matrix(Y_test, y_pred_tf > 0.5)
```

```python
print(classification_report(Y_test,y_pred_tf > 0.5))
```

## F1 스코어는 0.80

```python
precision_score(Y_test, y_pred_tf > 0.5)
```

```python
recall_score(Y_test, y_pred_tf > 0.5)
```

```python
f1_score(Y_test, y_pred_tf > 0.5)
```
