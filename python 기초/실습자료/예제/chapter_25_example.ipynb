## 따라해보기 예제  :   데코레이터로 클래스 갱신하기  

```python
def __init__(self,name,age) :
    self._name = name
    self._age = age
```

```python
class Descriptor :
    def __set_name__(self,Owner,name) :
        self._name = "_" + name
        
    def __get__(self, other, owner) :
        return other.__dict__[self._name]
        
    def __set__(self,other, value) :
        other.__dict__[self._name] = value
```

```python
def decorator(cls) :
    cls.__init__ = __init__

    return cls
```

```python
@decorator
class Klass:
    name = Descriptor()
    age = Descriptor()
```

```python
k = Klass("이름",45)
```

```python
k._name, k._age
```

```python
k.name, k.age
```
