## 따라해보기 예제  :  힙 자료구조

```python
class Heap :
    def __init__(self) :
        self._dict = {}
        
    def empty(self) :
        return True if len(self._dict) == 0 else False
    
    def pop(self, key) :
        if self._dict.get(key,0 ) :
            return self._dict.pop(key)
        else :
            return 0
        
    def push(self,key,value) :
        return self._dict.setdefault(key,value)
    
    def getHeap(self) :
        return self._dict.copy()
```

```python
h1 = Heap()
```

```python
h1.empty()
```

```python
h1.push("가을",100)
```

```python
h1.push("겨울",100)
```

```python
h1.getHeap()
```

```python
h1.pop("겨울")
```

```python
h1.getHeap()
```
