## 따라해보기 예제  :  순환문없이 순환돌리기

```python
def recusion(iterable) : 
    
    if not isinstance(iterable, list) :
        return "처리할 수 없습니다."
    
    if not hasattr(recusion,"result") :
        recusion.result = []
        
    if len(iterable) == 1 :
        recusion.result.append(iterable.pop())
        return None
        
    recusion.result.append(iterable.pop(0))
    return recusion(iterable)
```

```python
recusion([1,2,3,4])
```

```python
recusion.result
```

```python
recusion("문자열 가능여부")
```
