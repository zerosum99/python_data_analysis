## 따라해보기 예제  :   메소드 오버로딩   

```python
class  overload :
    def __init__(self,fget1=None,fget2=None, fget3=None) :
        self.fget1 = fget1
        self.fget2 = fget2
        self.fget3 = fget3
        
    def __get__(self,other,owner) :
        self._other = other
        return self
              
    def add(self,func) :
        if not self.fget2 :
            self.fget2 = func
        else :         
            if not self.fget3 :
                self.fget3 = func
        return self
    
    def __call__(self,*args) :
        if len(args) == 0 :
            print(self.fget1(self._other,*args))
        elif len(args) == 1 :
            print(self.fget2(self._other,*args))
        elif len(args) == 2 :
             print(self.fget3(self._other,*args))
        else :
            print("해당사항 없음")
        
```

```python
class Overload :

    
    def __init__(self,x,y) :
        self.x = x
        self.y = y
              
    @overload
    def add(self) :
        return self.x + self.y
    
    @add.add
    def add(self,z) :
        return self.x + self.y + z
    
    @add.add
    def add(self,a,b) :
        return self.x + self.y + a + b
```
