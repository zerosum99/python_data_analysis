## 따라해보기 예제  :   사용자 정의 메소드 클래스 만들기 

```python
class Method :
    def __init__(self,func) :
        self._func = func
        
    def __call__(self, *args, **kwargs) :
        return self._func(self,*args, **kwargs)
    
    def method(self) :
        return 100
```

```python
class A :
    @Method
    def a(self) :
        return 10
```

```python
A.__dict__
```

```python
bb = A()
```

```python
bb.a
```

```python
bb.a()
```

```python
bb.a.method()
```
