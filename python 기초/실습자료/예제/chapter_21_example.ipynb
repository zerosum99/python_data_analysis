## 따라해보기 예제  :   싱글턴 패턴  

```python
class Single :
    @staticmethod
    def __new__(cls,name,age) :
        print(" 객체를 하나 만듭니다.")
        if hasattr(cls, "_instance") :
            instance = cls._instance
        else :
            instance = object.__new__(cls)
            cls._instance = instance
        return instance
    
    def __init__(self,name,age) :
        print(" 객체의 속성을  초기화합니다.")
        if hasattr(type(self), "_init") :
            pass
        else : 
            self.name = name
            self.age = age
            type(self)._init = "1"
```

```python
a = Single("혼자만 있기", 20)
```

```python
b = Single("같이 있기", 30)
```
