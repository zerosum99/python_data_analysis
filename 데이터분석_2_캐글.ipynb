```python
df['Number of reported cases of cholera'] = df['Number of reported cases of cholera'].str.replace(" ", "")
```

```python
df['Number of reported cases of cholera'] = df['Number of reported cases of cholera'].astype('float')
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].str.replace("Unknown", "0")
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].str.replace("0 0", "0")
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].astype('float')
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].str.replace("Unknown", "0.0")
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].str.replace('0.0 0.0', "0.0")
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].astype('float')
```
